## 1.2. Kursa pārskats: refleksija par W1–W3

Visvieglākais no trim ML tipiem man bija **klasifikācija**, jo jautājums 
"vai klients pirks?" ir intuitīvs — atbilde ir vai nu jā, vai nē. 
F1 metrika arī bija viegli saprotama, jo tā ņem vērā gan precizitāti, 
gan atcerēšanos vienā skaitlī.

Visgrūtākais bija **regresija** — īpaši PageValues datu gadījumā, kur R² = 0.046 
nozīmē, ka modelis izskaidro tikai 4.6% no datu variācijas. Galvenā problēma 
bija datu sadalījums: lielākā daļa vērtību bija 0, un tikai daži klienti 
ģenerēja augstu PageValues. Šādos apstākļos pat labākais modelis cīnās.

**Klasterizācija** bija interesanta, jo nav vienas "pareizas" atbildes — 
K=3 izvēlējāmies pēc Elbow metodes, bet interpretācija (ko nozīmē katrs 
klasteris?) prasīja domāšanu, nevis tikai skaitļu salīdzināšanu.

Ja sāktu no sākuma, **vairāk laika veltītu datu izpētei (EDA)** pirms 
modeļa veidošanas. W2 gadījumā PageValues problēma bija redzama jau 
datu sadalījumā — ja to būtu pamanījusi agrāk, varētu izvēlēties 
atšķirīgu pieeju vai citu target kolonnu.

**Intuitīvākā metrika** man bija F1 score, jo to var izskaidrot vienkārši: 
"cik bieži modelis ir pareizs, kad saka 'pirks', un cik daudz pircēju 
tas pamana?" R² savukārt ir abstraktāks — 0.046 nozīmē maz, bet cik 
maz — tas prasa pieredzi, lai novērtētu.

**Data leakage** mūsu darbā nebija kritiska problēma, jo izmantojām 
Pipeline, kas nodrošina, ka StandardScaler tiek trenēts tikai uz 
treniņa datiem. Overfitting risks bija lielāks polinomiālajai regresijai, 
kur Train R² un Test R² atšķirība liecināja par pārmācīšanos.

## 2.1. Kādā situācijā lietot kuru pieeju?

**Klasifikāciju** lieto, kad atbilde ir kategorija. Piemēram, medicīnā: 
"Vai pacientam ir diabēts?" — atbilde ir jā vai nē. Modelis mācās no 
pacientu analīžu datiem un prognozē diagnozi.

**Regresiju** lieto, kad atbilde ir skaitlis. Piemēram, izglītībā: 
"Cik punktu skolēns iegūs eksāmenā?" — atbilde ir konkrēts skaitlis. 
Modelis var ņemt vērā apmeklētību, mājas darbu izpildi un iepriekšējos rezultātus.

**Klasterizāciju** lieto, kad nav zināmas grupas un vēlamies tās atklāt. 
Piemēram, bibliotēkā: "Kādi lasītāju tipi mums ir?" — nav pareizās atbildes, 
bet K-Means var atklāt, ka ir "bieži apmeklētāji", "sezonālie lasītāji" 
un "digitālie lietotāji".

**Pipeline un Cross-validation** ir nepieciešami visos gadījumos, jo 
Pipeline novērš data leakage (skalers nemācās no testa datiem), bet 
Cross-validation dod uzticamu rezultātu — nevis viena nejaušā split 
vērtību, bet vidējo no 5 dažādiem testiem.

# 5. uzdevums: Bāzes modelis (Baseline)

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import numpy as np

# Sagatavo X un y
X = df.drop(columns=['Target'])
le = LabelEncoder()
y = le.fit_transform(df['Target'])

print(f"Klašu kodējums: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# Bāzes Pipeline
pipe_baseline = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)

# Cross-validation
cv_scores = cross_val_score(pipe_baseline, X, y, cv=5, scoring='f1_weighted')

print(f"\nBāzes modelis: Logistic Regression")
print(f"CV F1 (weighted) pa fold'iem: {cv_scores.round(3)}")
print(f"Vidējais CV F1: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
```

## 5.2. Bāzes rezultāta interpretācija

Bāzes modelis (Logistic Regression) sasniedza CV F1 = **0.749 ± 0.009**. 

Tas ir labs sākuma rezultāts — modelis jau spēj pareizi klasificēt

# 6. uzdevums: Feature Engineering un datu sagatavošana

```python
# 6.3. Jaunu pazīmju izveide
df_fe = df.copy()

# Jaunā pazīme 1: kopējais apstiprinātais kursu skaits (1. + 2. semestris)
df_fe['Total_approved'] = (df_fe['Curricular units 1st sem (approved)'] + 
                           df_fe['Curricular units 2nd sem (approved)'])

# Jaunā pazīme 2: apstiprināšanas līmenis (approved / enrolled)
df_fe['Approval_rate'] = (df_fe['Total_approved'] / 
                          (df_fe['Curricular units 1st sem (enrolled)'] + 
                           df_fe['Curricular units 2nd sem (enrolled)'] + 0.001))

# Jaunā pazīme 3: vidējā atzīme abos semestros
df_fe['Avg_grade'] = ((df_fe['Curricular units 1st sem (grade)'] + 
                       df_fe['Curricular units 2nd sem (grade)']) / 2)

print("Jaunās pazīmes izveidotas!")
print(f"\nTotal_approved statistika:\n{df_fe['Total_approved'].describe().round(2)}")
print(f"\nApproval_rate statistika:\n{df_fe['Approval_rate'].describe().round(2)}")
print(f"\nAvg_grade statistika:\n{df_fe['Avg_grade'].describe().round(2)}")

# Korelācija ar Target
X_fe = df_fe.drop(columns=['Target'])
print(f"\nDataseta izmērs pēc feature engineering: {X_fe.shape}")
```

## 6.4. Feature Engineering skaidrojums

Izveidotas **3 jaunas pazīmes**, kas apvieno informāciju no abiem semestriem:

**Total_approved** — kopējais apstiprinātais kursu skaits (1. + 2. semestris). 
Vidēji students apstiprina 9 kursus. Studenti ar 0 apstiprinātiem kursiem 
visticamāk ir Dropout kategorijā.

**Approval_rate** — apstiprināšanas līmenis (cik % no reģistrētajiem kursiem 
students sekmīgi nokārtoja). Vidēji 68%. Šī pazīme ir informatīvāka par 
absolūto skaitu, jo normalizē kursu slodzi.

**Avg_grade** — vidējā atzīme abos semestros kopā. Vidēji 10.44 no 20. 
Šī pazīme apvieno divus semestrus vienā skaitlī, samazinot dimensionalitāti.

Visas trīs pazīmes ir loģiski saistītas ar dropout risku — students, 
kurš neapstiprina kursus un saņem zemas atzīmes, visticamāk pamet studijas.

# 7. uzdevums: Modeļu salīdzināšana un hiperparametru optimizācija

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Sagatavo datus ar jaunajām pazīmēm
X_fe = df_fe.drop(columns=['Target'])
y = le.fit_transform(df_fe['Target'])

# Modeļu saraksts
models = {
    'Logistic Regression (baseline)': make_pipeline(
        StandardScaler(), 
        LogisticRegression(max_iter=1000, random_state=42)),
    'Random Forest': make_pipeline(
        StandardScaler(), 
        RandomForestClassifier(n_estimators=100, random_state=42)),
    'Gradient Boosting': make_pipeline(
        StandardScaler(), 
        GradientBoostingClassifier(n_estimators=100, random_state=42))
}

# Salīdzina visus modeļus
rezultati_modeli = {}
for name, pipe in models.items():
    scores = cross_val_score(pipe, X_fe, y, cv=5, scoring='f1_weighted')
    rezultati_modeli[name] = scores
    print(f"{name}:")
    print(f"  CV F1: {scores.mean():.3f} ± {scores.std():.3f}")
    print()
```

```python
from sklearn.model_selection import GridSearchCV

# Labākais modelis — Gradient Boosting
pipe_grid = make_pipeline(
    StandardScaler(),
    GradientBoostingClassifier(random_state=42)
)

param_grid = {
    'gradientboostingclassifier__n_estimators': [100, 200],
    'gradientboostingclassifier__max_depth': [3, 5],
    'gradientboostingclassifier__learning_rate': [0.05, 0.1]
}

grid_search = GridSearchCV(
    pipe_grid, param_grid, 
    cv=5, scoring='f1_weighted', 
    n_jobs=-1, verbose=1
)

grid_search.fit(X_fe, y)

print(f"Labākie parametri: {grid_search.best_params_}")
print(f"Labākais CV F1: {grid_search.best_score_:.3f}")
```

```python
# Salīdzinājuma tabula
tabula = pd.DataFrame({
    'Modelis': [
        'Bāzes modelis (Logistic Regression)',
        'Random Forest',
        'Gradient Boosting',
        'GridSearchCV (Gradient Boosting)'
    ],
    'CV F1 (vidējais)': [0.748, 0.766, 0.775, 0.775],
    'CV F1 (std)': [0.011, 0.005, 0.011, 0.011],
    'Uzlabojums vs bāzi': ['-', '+0.018', '+0.027', '+0.027']
})
print(tabula.to_string(index=False))
```

## 7.3. Modeļu salīdzinājums — secinājumi

| Modelis | CV F1 | Std |
|---|---|---|
| Logistic Regression (bāze) | 0.748 | ±0.011 |
| Random Forest | 0.766 | ±0.005 |
| Gradient Boosting | 0.775 | ±0.011 |
| GridSearchCV labākais | 0.775 | ±0.011 |

**Gradient Boosting** ir labākais modelis ar F1 = 0.775 — par +0.027 
labāks nekā bāzes modelis. GridSearchCV apstiprināja, ka noklusējuma 
parametri (learning_rate=0.1, max_depth=3, n_estimators=100) jau ir 
optimāli šiem datiem. Random Forest ir stabilāks (std=0.005), bet 
Gradient Boosting sasniedz augstāku precizitāti.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Apmācām labāko modeli
best_model = grid_search.best_estimator_
X_train, X_test, y_train, y_test = train_test_split(
    X_fe, y, test_size=0.2, random_state=42, stratify=y)

best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

# 8.1. Feature Importance
feature_names = X_fe.columns.tolist()
importances = best_model.named_steps['gradientboostingclassifier'].feature_importances_
indices = np.argsort(importances)[::-1][:15]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Feature importance grafiks
axes[0].barh([feature_names[i] for i in indices[::-1]], 
             importances[indices[::-1]], color='#3498db')
axes[0].set_title('Top 15 svarīgākās pazīmes')
axes[0].set_xlabel('Importance')

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred,
    display_labels=le.classes_,
    ax=axes[1],
    colorbar=False,
    cmap='Blues'
)
axes[1].set_title('Confusion Matrix')

plt.tight_layout()
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
